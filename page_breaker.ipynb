```python
import pandas as pd

def find_markers(file_name, start_marker_content, start_marker_col, end_marker_content, end_marker_col):
    # Load the excel file
    with pd.ExcelFile(file_name) as xls:
        df = xls.parse('总表')
        
        # Convert column letters to 0-based column indices
        start_marker_col = ord(start_marker_col.upper()) - ord('A')
        end_marker_col = ord(end_marker_col.upper()) - ord('A')

        # Find start marker rows
        start_rows = df[df.iloc[:, start_marker_col].apply(lambda x: start_marker_content in str(x))].index.tolist()

        # Find end marker rows
        end_rows = df[df.iloc[:, end_marker_col].apply(lambda x: end_marker_content in str(x))].index.tolist()

    # 调整开始标记和结束标记
    # 开始标记行+1
    start_rows = [x+1 for x in start_rows]
    end_rows = [x+3 for x in end_rows]

    # Return results
    return start_rows, end_rows, len(start_rows), len(end_rows)

# Example usage:
start_marker_content = "标段"
start_marker_col = 'K' 
end_marker_content = "转至"
end_marker_col = 'I'

start_rows, end_rows, num_start, num_end = find_markers('户内装修A-G标段合并.xlsx', start_marker_content, start_marker_col, end_marker_content, end_marker_col)

print(start_rows)
print(end_rows)
print(num_start)
print(num_end)

if num_start != num_end:
    print(f'错误：开始标记和结束标记的数量不一致！')
    print(f'开始标记数量：{num_start}，结束标记数量：{num_end}。')
else:
    print(f'开始标记和结束标记的数量一致。')
```

```python
# 为VBA格式对数组进行换行 
def format_for_vba(nums, numbers_per_line=5):
    # Split the numbers into groups
    groups = [nums[i:i+numbers_per_line] for i in range(0, len(nums), numbers_per_line)]
    
    # Format each group into a line of VBA code
    lines = [", ".join(map(str, group)) + " _" for group in groups]
    
    # Join all lines into a single string
    result = "\n".join(lines)
    return result

print(format_for_vba(end_rows))
```

```python
import pandas as pd

# 文件路径
file_path = 'test.xlsx'
output_path = 'test_output.xlsx'
start_col = ord('B') - ord('A')
end_col = ord('K') - ord('A')

start_rows = [x - 1 for x in start_rows]
end_rows = [x - 1 for x in end_rows]
# 读取excel文件
df = pd.read_excel(file_path, engine='openpyxl', sheet_name='总表')
print(df.columns)
print(f'df.columns.get_loc(start_col):{df.columns.get_loc(start_col)}')
print(f'df.columns.get_loc(end_col):{df.columns.get_loc(end_col)}')

# 初始化一个空的DataFrame用于存储所有选择的数据部分
selected_data = pd.DataFrame()

# 循环遍历每一对start_row和end_row来提取数据
for sr, er in zip(start_rows, end_rows):
    temp_data = df.iloc[sr:er+1, df.columns.get_loc(start_col):df.columns.get_loc(end_col)+1]
    selected_data = pd.concat([selected_data, temp_data])

# 保存选择的数据到新的excel文件
selected_data.to_excel(output_path, index=False, engine='openpyxl')

print(f"数据已保存到 {output_path}")
```

```python
from openpyxl import load_workbook
from openpyxl.worksheet.pagebreak import Break

def set_print_ranges(file_name, start_rows, end_rows, start_col='B', end_col='K'):
    # Load the workbook and get the active worksheet
    wb = load_workbook(file_name)
    ws = wb.active

    # Set the print areas for each range
    areas = []
    for start_row, end_row in zip(start_rows, end_rows):
        areas.append(f"{start_col}{start_row}:{end_col}{end_row}")
    
    # Join the areas with commas to set multiple print areas
    ws.page_setup.print_area = ",".join(areas)

    # Save the workbook with the updated print areas
    wb.save(file_name)

# Using the previously found start_rows and end_rows
set_print_ranges('户内装修A-G标段合并.xlsx', start_rows, end_rows)
```

```python
from openpyxl import load_workbook
from openpyxl.worksheet.pagebreak import RowBreak
def set_print_pages_with_cols(file_name, start_rows, end_rows, start_col='B', end_col='K'):
    # Load the workbook and get the active worksheet
    wb = load_workbook(file_name)
    ws = wb.active

    # Set the column range for printing
    col_start_idx = ord(start_col.upper()) - ord('A')
    col_end_idx = ord(end_col.upper()) - ord('A')
    ws.print_options = {'gridLines': False, 'horizontalCentered': True, 'verticalCentered': True}
    ws.print_area = f"{start_col}1:{end_col}{ws.max_row}"

    # For each range, add a page break after the end_row
    for end_row in end_rows:
        brk = RowBreak()
        brk.id = end_row + 1
        ws.row_breaks.append(brk)

    # Save the workbook with the page breaks and column range added
    wb.save(file_name)

# Using the previously found start_rows and end_rows
set_print_pages_with_cols('户内装修A-G标段合并.xlsx', start_rows, end_rows)
```

```python
from openpyxl import load_workbook
from openpyxl.worksheet.pagebreak import RowBreak
def set_print_pages_with_corrected_breaks(file_name, start_rows, end_rows, start_col='B', end_col='K'):
    # Load the workbook and get the active worksheet
    wb = load_workbook(file_name)
    ws = wb.active

    # Set the column range for printing
    col_start_idx = ord(start_col.upper()) - ord('A')
    col_end_idx = ord(end_col.upper()) - ord('A')
    
    ws.print_area = f"{start_col}1:{end_col}{ws.max_row}"

    # Set print options
    ws.print_options.gridLines = False
    ws.print_options.horizontalCentered = True
    ws.print_options.verticalCentered = True

    # Set page orientation and fit the content
    ws.page_setup.orientation = ws.ORIENTATION_LANDSCAPE
    ws.page_setup.fitToHeight = 0
    ws.page_setup.fitToWidth = 1

    # Clear any existing page breaks
    ws.row_breaks = []

    # For each range, add a page break after the end_row
    for end_row in end_rows:
        brk = RowBreak()
        brk.id = end_row + 1
        ws.row_breaks.append(brk)

    # Save the workbook with the page breaks and column range added
    wb.save(file_name)

# This is the updated function. You would still need to run it in your environment with the proper arguments.
set_print_pages_with_corrected_breaks('户内装修A-G标段合并.xlsx', start_rows, end_rows)
```
